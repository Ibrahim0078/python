```python
#1.	Which columns have the wrong data type (e.g., time and releasedate as strings) and which should be converted?
df['price'] = (df['price'].str.replace(r'[\$,]', '', regex=True).replace('', pd.NA).pipe(pd.to_numeric, errors='coerce'))
df['releasedate'] = pd.to_datetime(df['releasedate'], errors='coerce')
df.info()
df.head()
```

```python
#2.	Are there any rows or columns that are entirely empty or nearly empty — should any be dropped?
df.isna().sum()
```

```python
#3.	Are column names consistent (lowercase, underscores)? Should you rename releasedate → release_date, stars → rating_text, etc.?
df = df.rename(columns={
    "releasedate": "release_date",
    "stars": "rating_text",
})
```

```python
df.head()
```

```python
#4.	How can you remove the prefixes Writtenby: and Narratedby: from author and narrator and store only the names?
df['author'] = df['author'].str.replace(r'^Writtenby:\s*', '', regex=True)
df['narrator'] = df['narrator'].str.replace(r'^Narratedby:\s*', '', regex=True)
df.head()
```

```python
#5.	Can you split name into parts (e.g., series name vs. book numbers, or title vs subtitle) where appropriate?
df[['title', 'subtitle']] = df['name'].str.split(':', n=1, expand=True)
df.head()
```

```python
#6.	How can you parse time into consistent numeric fields like duration_minutes (or duration_hours + duration_minutes)?
df['hours'] = df['time'].str.extract(r'(\d+)\s*hrs?').fillna(0).astype(int)
df['minutes'] = df['time'].str.extract(r'(\d+)\s*mins?').fillna(0).astype(int)
df.head()
```

```python
#7.	How can you parse stars to extract a numeric rating (e.g., 4.5) and the number of ratings (e.g., 34)? What about when spacing or words are missing?
# Extract numeric rating
df['rating_count'] = (df['rating_text'].str.extract(r'([\d,]+)\s*(?=ratings\b)', expand=False).str.replace(',', '', regex=False).astype('Int64'))

# Extract number of ratings
df['rating'] = (df['rating_text'].astype('string').str.extract(r'(\d+(?:\.\d+)?)', expand=False).astype(float))
df.head()
```

```python
#8.	Are releasedate values in a consistent format (mm/dd/yyyy vs dd/mm/yyyy)? How can you programmatically detect the format and convert to datetime64[ns]?
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df.head()
```

```python
#9.	After converting, can you extract useful features like release_year, release_month, release_day, or release_weekday?
df['release_year'] = df['release_date'].dt.year
df['release_month'] = df['release_date'].dt.month
df['release_day'] = df['release_date'].dt.day
df['release_weekday'] = df['release_date'].dt.dayofweek
df.head(2)
```

```python
#10.	Are there impossible or invalid dates (e.g., month > 12 if format misinterpreted)? How will you detect and fix them?
df['release_date'].isna().sum()
```

```python
#11.	Which values should be considered "missing" (e.g., "", "NA", "None", "-", "n/a") and how will you standardize them to np.nan?
import numpy as np

missing_vals = ["", "NA", "None", "-", "n/a", "N/A"]
df = df.replace(missing_vals, np.nan)
```

```python
#12.	For each column, how many missing values exist and are there patterns (missing by author, by year, etc.)?
report = pd.DataFrame({
    'dtype': df.dtypes.astype(str),        # data type of each column
    'n_missing': df.isna().sum(),          # number of missing values
    'pct_missing': df.isna().mean()*100,   # percentage missing
    'n_unique': df.nunique(dropna=False)   # how many unique values
}).sort_values('pct_missing', ascending=False)

print(report)
# Missing price by release year (%)
abc = df.groupby('release_year')['price'].apply(lambda x: x.isna().mean()*100).sort_values(ascending=False).head(10)
# Missing price by author (%)
defg = df.groupby('author')['price'].apply(lambda x: x.isna().mean()*100).sort_values(ascending=False).head(10)
print(abc)
print(defg)
```

```python
#13.	For columns with missing values, which imputation strategy is appropriate (leave as missing, fill with mode/median, forward/backfill, or domain default)?
df['price'] = df['price'].fillna(df['price'].median())
df = df.dropna(subset=['title'])
df = df.drop(columns=['subtitle', 'rating', 'rating_count'])
df.head()
```

```python
#14.	Are there leading/trailing spaces or extra internal whitespace in any text fields? How will you trim them?
df['author'] = df['author'].str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
#15.	Does casing matter? Should you standardize language, author, and narrator to title case or lowercase?
df['language'] = df['language'].str.lower()
df['author'] = df['author'].str.lower()
df['narrator'] = df['narrator'].str.lower()
df['title'] = df['title'].str.lower()
```

```python
#16.	Are there punctuation or special characters to remove (e.g., stray colons, HTML entities)?
df['author'] = df['author'].str.replace(r"[^\w\s]", "", regex=True)
```

```python
#17.	Do you need to correct common misspellings or inconsistent variants (e.g., Englis, english, ENGLISH)?
df['language'] = df['language'].replace({
    'englis': 'english',
    'ENGLISH': 'english',
    'Englsh': 'english'
})
```

```python
#18.	Is price stored as numeric or string? If string, how will you strip currency symbols and thousands separators then convert to float?
df['price'] = (df['price'].astype('string').str.replace(r'[\$,]', '', regex=True).replace('', pd.NA).pipe(pd.to_numeric, errors='coerce'))
```

```python
#20.	After extracting rating (numeric) and rating_count (integer), are there any non-numeric or malformed values to handle?
df['rating_text'].value_counts(dropna=False)
df.head()
```

```python
#21.	Should author and narrator names be normalized to a consistent format (e.g., First Last)? How to handle initials, middle names, etc.?
df['author'] = df['author'].str.strip().str.title()
df['narrator'] = df['narrator'].str.strip().str.title()
df['author'] = df['author'].str.replace(r'\s+', ' ', regex=True)   # collapse spaces
df['author'] = df['author'].str.replace(r'\.\s*', '.', regex=True) # fix periods
df['author'] = df['author'].str.replace(',', '', regex=False)
print("Unique authors:", df['author'].nunique())
print("Unique narrators:", df['narrator'].nunique())
df.head()
```
