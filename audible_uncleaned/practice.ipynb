```python
import pandas as pd
df = pd.read_csv("C:\\Users\\ibrah\\Python\\audible_uncleaned\\audible_uncleaned.csv")
```

```python
df.head(2)
```

```python
# Remove 'Writtenby:' and 'Narratedby:' + fix spacing
df['author'] = df['author'].str.replace("Writtenby:", "").str.strip()
df['narrator'] = df['narrator'].str.replace("Narratedby:", "").str.strip()

df.head(2)
```

```python
hours = df['time'].str.extract(r'(\d+)\s*hrs?').fillna(0).astype(int)
minutes = df['time'].str.extract(r'(\d+)\s*mins?').fillna(0).astype(int)
df["total_minutes"] = hours * 60 + minutes
df.head(2)
```

```python
df['releasedate'] = pd.to_datetime(df['releasedate'], errors='coerce')
df.head(2)
```

```python
# Extract the rating (e.g., 4.5, 5, etc.)
df['rating'] = df['stars'].str.extract(r'(\d+\.?\d*)').astype(float)

# Extract the number of ratings (digits at the end)
df['rating_count'] = df['stars'].str.extract(r'(\d+)\s*ratings').fillna(0).astype(int)

# Check result
df.head(2)
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.head(2)
```

```python
print(df.isna().sum())
```

```python
missing_percent = df.isna().mean() * 100
print(missing_percent)
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
print(df.isna().sum())
```

```python
df.drop(columns=['releasedate'], inplace=True)
df['price'] = df['price'].fillna(df['price'].median())
```

```python
print(df.isna().sum())
```

```python
df['language'] = df['language'].str.capitalize()
```

```python
print(df.dtypes)
print(df.describe())
print(df['language'].unique())
print(df['author'].nunique(), "unique authors")
print("Unique narrators:", df['narrator'].nunique())
```

```python
df.drop(columns=['stars'], inplace=True)
df.drop(columns=['time'], inplace=True)
```
