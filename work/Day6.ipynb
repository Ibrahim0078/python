```python
class student:
    def __init__ (self, name, fatherName, roll_number, marks, address, contactNumber, email):
        self.name = name
        self.fatherName = fatherName
        self.rollNumber = roll_number
        self.marks = marks
        self.__address = address
        self.__contactNumber = contactNumber
        self.__email = email
    def updateAddress(self, address):
        self.__address = address
    def updateContactNumber(self, contactNumber):
        self.__contactNumber = contactNumber
    def updateEmail(self, email):
        self.__email = email
    def updateMarks(self, marks):
        self.marks = marks
    def displayFullInfo(self):
        print(f"student name is {self.name}\nfather name is {self.fatherName}\nroll number is {self.rollNumber}\nmarks are {self.marks}\nAddress is {self.__address}\nPhone Number is {self.__contactNumber}\nEmail is {self.__email}")
    def displayShortInfo(self):
        print(f"student name is {self.name}\nfather name is {self.fatherName}\nroll number is {self.rollNumber}")
    def contactInfo(self):
        print(f"{self.name}'s contact info is:\nAddress is {self.__address}\nPhone Number is {self.__contactNumber}\nEmail is {self.__email}")
    def hasPassed(self):
        if self.marks <= 40:
            print(f"The student {self.name} has failed")
        else:
            print(f"The student {self.name} has passed")
    def isScholarshipEligible(self):
        if self.marks <= 85:
            print(f"The student {self.name} is not eligible for scholership")
        else:
            print(f"The student {self.name} is eligible for scholership")
class newStudent(student):
    pass
s1 = newStudent("Ali Khan", "Ahmed Khan", 101, 92, "Lahore, Pakistan", "03001234567", "alikhan@gmail.com")
s2 = newStudent("Sara Malik", "Rashid Malik", 102, 76, "Karachi, Pakistan", "03119876543", "saramalik@yahoo.com")
s3 = newStudent("Usman Ahmed", "Khalid Ahmed", 103, 38, "Islamabad, Pakistan", "03334567891", "usman.ahmed@hotmail.com")

# Function Calls
s1.displayFullInfo()
s1.hasPassed()
s1.isScholarshipEligible()
print("---- Updating Ali's Contact ----")
s1.updateContactNumber("03009998888")
s1.contactInfo()
print("\n==========================\n")
print("\n**************************\n")
print(s1.name)
#print(s1.__address)

print("\n==========================\n")

s2.displayShortInfo()
s2.hasPassed()
s2.isScholarshipEligible()

print("\n==========================\n")

s3.displayFullInfo()
s3.hasPassed()
s3.isScholarshipEligible()
print("---- Updating Usman's Marks ----")
s3.updateMarks(55)
s3.hasPassed()
```
