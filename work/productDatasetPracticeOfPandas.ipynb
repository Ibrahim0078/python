```python
#2.	Add a new column: average price per category (using transform).
df["average_price_per_category"] = df.groupby("category")["unit_price"].transform("mean")
```

```python
df.head()
```

```python
#3.	Add a column: difference between each product’s price and its category average.
df["diff_between_product"] = df["unit_price"] - df["average_price_per_category"]
```

```python
df.head(2)
```

```python
#4.	Add a column: % contribution of each product’s stock to its category stock.
# Calculate total stock per category
category_totals = df.groupby("category")["stock"].transform("sum")
# Add the % contribution column
df["stock_pct_of_category"] = (df["stock"] / category_totals) * 100
df.head(2)
```

```python
#5.	Rank products within each category by price.
df["price_rank_in_category"] = df.groupby("category")["unit_price"].rank(method="dense", ascending=False)
df.head(2)
```

```python
# 1. Create a summary DataFrame with columns: 
# Category, Total_Products, Avg_Price, Total_Stock, Min_Price, Max_Price
df.groupby("category").agg(Total_Products=("name", "count"),Avg_Price=("unit_price", "mean"),Total_Stock=("stock", "sum"),Min_Price=("unit_price", "min"),Max_Price=("unit_price", "max")).reset_index()
```

```python
# 2. Find the ratio of in_stock to pre_order products per category
availability_ratio = (
    df.groupby(["category", "availability"]).size().unstack(fill_value=0)
)
availability_ratio["in_stock_to_pre_order"] = (
    availability_ratio["in_stock"] / availability_ratio["pre_order"]
)
print(availability_ratio)
print(availability_ratio["in_stock_to_pre_order"])
```

```python
# 3. For each brand, compute number of products, total stock, average price
df.groupby("brand").agg(
    num_products=("name", "count"),
    total_stock=("stock", "sum"),
    avg_price=("unit_price", "mean")
).reset_index()
```

```python
# 4. Create a column "High Stock" if Stock > 500, otherwise "Low Stock"
df["stock_level"] = df["stock"].apply(lambda x: "High Stock" if x > 500 else "Low Stock")
df.head(2)
```
